```python
#Sentinel-2 SR Harmonized Daten verfügbar von 2017-03-28 bis 2026-01-25

collection_S2 = ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED")
```

```python
#For-Schleifen erstellen, die mir zeigen, ob es für die jeweilige AOI Daten von Sentinel-2 gibt

#Schleife für Bilder aus 2017-2018 mit weniger als 10% Wolken:

for name, boundary in boundaries.items():
    S2_1 = (
        collection_S2
        .filterBounds(boundary)
        .filterDate("2017-03-28", "2018-12-31")
        .filter(ee.Filter.lte("CLOUDY_PIXEL_PERCENTAGE", 10))
    )

    exists = S2_1.size().gt(0).getInfo()

    if exists:
        print(f"{name}: Bild vorhanden")

    else:
        print(f"{name}: Kein Bild")
```

```python
#For-Schleife erstmal ohne Wolken (für Daten aus 2017/18) und dann bestes Bild anzeigen lassen (weil nicht für jede AOI ein Bild von 10% vorhanden)

for name, boundary in boundaries.items():
    S2_1_Wolken = (
        collection_S2
        .filterBounds(boundary)
        .filterDate("2017-03-28", "2018-12-31")
    )

    exists = S2_1_Wolken.size().gt(0)

    if exists.getInfo():
        best_image = S2_1_Wolken.sort("CLOUDY_PIXEL_PERCENTAGE").first()
        cloud_value = best_image.get("CLOUDY_PIXEL_PERCENTAGE").getInfo()
        print(f"{name} Wolkenanteil: {cloud_value}%")
    else:
        print(f"{name}: Kein Bild")
```

```python
#Schleife für Bilder aus 2025:

for name, boundary in boundaries.items():
    S2_2 = (
        collection_S2
        .filterBounds(boundary)
        .filterDate("2025-01-01", "2025-12-31")
        .filter(ee.Filter.lte("CLOUDY_PIXEL_PERCENTAGE", 10))
    )

    exists = S2_2.size().gt(0).getInfo()

    if exists:
        print(f"{name}: Bild vorhanden")

    else:
        print(f"{name}: Kein Bild")
```

```python
#!!EXPORT!! 

#Bildqualität an sich noch oft sehr schlecht - klappt also noch nicht so gut
#vlt muss auch RGB irgendwie angepasst werden oder so
#Wegen den teils zu großen Dateien noch versuchen entweder in Google Drive zu exportieren oder:
#best_image_rgb = best_image.select(["B4", "B3", "B2"])


#Schleife für Bilder aus 2017-2018 mit möglichst wenig Wolken:

for name, boundary in boundaries.items():
    S2_1 = (
        collection_S2
        .filterBounds(boundary)
        .filterDate("2017-03-28", "2018-12-31")
    )

    if S2_1.size().getInfo() > 0:
        #wenn es ein Bild gibt, dann das mit den wenigsten Wolken auswählen
        best_image = S2_1.sort("CLOUDY_PIXEL_PERCENTAGE").first()

        #Bänder auswählen, damit Export klappt (und für NDWI später)
        export_image = best_image.select(['B3','B8']).toFloat()

        #Export als GeoTIFF
        geemap.ee_export_image_to_drive( #ohne _to_drive für lokal
            export_image,
            #filename=f"C:/Users/Windows/Desktop/Master/GIS1/Bilder_S2_1/{name}_image.tif", #falls lokal gespeichert werden soll, sonst lieber Drive:
            description = f"S2_1_{name}",
            folder="TIFFs_S2_1",
            scale=10,         #10m Auflösung für RGB/Bänder wegen Sentinel 2
            region=boundary, #schneidet nur die AOI raus
            crs="EPSG:32649"  #EPSG für sc Meer
        )
        print(f"{name}: Bild gefunden und exportiert")

    else:
        print(f"{name}: Kein Bild gefunden")
```

```python
#!!EXPORT!!


#Schleife für Bilder aus 2025 mit möglichst wenig Wolken:

for name, boundary in boundaries.items():
    S2_2 = (
        collection_S2
        .filterBounds(boundary)
        .filterDate("2025-01-01", "2025-12-31")
        .filter(ee.Filter.lte("CLOUDY_PIXEL_PERCENTAGE", 10))
    )

    if S2_2.size().getInfo() > 0:
        #wenn es ein Bild gibt, dann das mit den wenigsten Wolken auswählen
        best_image = S2_2.sort("CLOUDY_PIXEL_PERCENTAGE").first()

        #Export als GeoTIFF
        geemap.ee_export_image(
            best_image,
            filename=f"C:/Users/Windows/Desktop/Master/GIS1/Bilder_S2_2/{name}_image.tif",
            scale=10,         #10m Auflösung für RGB/Bänder wegen Sentinel 2
            region=boundary, #schneidet nur die AOI raus
            crs="EPSG:4326"  #??? ggf. ändern
        )
        print(f"{name}: Bild gefunden und exportiert")

    else:
        print(f"{name}: Kein Bild gefunden")
```
